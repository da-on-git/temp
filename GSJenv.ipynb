###Calculate Factorial using a for loop

```python
def factorialA(n):
  fact = 1
  for i in range(1, n+1):
    fact = fact * i
  return fact
```

###Calculate Factorial using recursion

```python
def factorialB(n):
  if n == 1:
    fact = 1
  else:
    fact = n * factorialB(n-1)
  return fact
```

```python
%timeit factorialA(100)
```

```python
%timeit factorialB(100)
```
